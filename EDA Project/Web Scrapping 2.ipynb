```python
Brand=[]
Strap=[]
Price=[]
Original_Price=[]
Discount= []
Deals=[]

for i in range(1,31):
    url='https://www.flipkart.com/search?q=smartwatch&sid=ajy%2Cbuh&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_8_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_8_na_na_na&as-pos=1&as-type=RECENT&suggestionId=smartwatch%7CSmart+Watches&requestId=0d0810b6-0413-449a-bf26-d54d847d4d52&as-backfill=on&page=20'    
    page=requests.get(url)
    soup=BeautifulSoup(page.text)
    container=soup.find_all('div' ,class_="_2B099V")
    for i in container:
        text=i.find('a' ,class_="IRpwTa")
        if text:
            Brand.append(text.text)
        else:
            Brand.append(np.nan)
            
        
        text=i.find('div' ,class_="_3I9_wc")
        if text:
            Original_Price.append(text.text)
        else:
            Original_Price.append(np.nan)
    
        
        text=i.find('div' ,class_="_3Ay6Sb")
        if text:
            Discount.append(text.text)
        else:
            Discount.append(np.nan)
            
        
        text=i.find('div',class_="_30jeq3")
        if text:
            Price.append(text.text)
        else:
            Price.append(np.nan)
        
        
        text=i.find('div' ,class_="_3eWWd-")
        if text:
            Strap.append(text)
        else:
            Strap.append(np.nan)
        
        text=i.find('div' ,class_="_2Tpdn3 _18hQoS" )
        if text:
            Deals.append(text)
        else:
            Deals.append(np.nan)
```

```python
print(len(Brand))
print(len(Price))
print(len(Original_Price))
print(len(Discount))
print(len(Strap))
```

```python
Smart_Watch={"Brand":Brand,"Price":Price,"MRP":Original_Price,"Discount":Discount,"Strap":Strap,'Deals':Deals}
```

```python
Smart_Watch=pd.DataFrame(Smart_Watch)
df=Smart_Watch
```
