```python
unique_brands=df['Brand'].unique()
bar_width = 0.35
x = np.arange(len(unique_brands))

plt.figure(figsize=(15, 8))


plt.bar(x - bar_width/2, df.groupby('Brand')['MRP'].mean(), width=bar_width, label='MRP')
plt.bar(x + bar_width/2, df.groupby('Brand')['Price'].mean(), width=bar_width, label='Price')


plt.xlabel('Brand')
plt.ylabel('MRP & Price')
plt.title('MRP and Price by Brand')
plt.xticks(x,unique_brands)
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
